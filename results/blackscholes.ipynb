```python
# blackscholes_pthreads = pd.read_csv("blackscholes/small/benchmark-gcc-pthreads-55.out")[:54]
# blackscholes_ff = pd.read_csv("blackscholes/small/benchmark-gcc-ff-55.out")[:54]
# blackscholes_openmp = pd.read_csv("blackscholes/small/benchmark-gcc-openmp-55.out")[:54]
# blackscholes_tbb = pd.read_csv("blackscholes/small/benchmark-gcc-tbb-55.out")[:54]
# blackscholes_skepu_openmp = pd.read_csv("blackscholes/small/benchmark-gcc-skepu-openmp-55.out")[:54]
# blackscholes_hybrid = pd.read_csv("blackscholes/small/benchmark-gcc-skepu-hybrid.out")[:54]
# serial_time = 0.097964
# serial_energy = 3.75711
# label = "small"

# blackscholes_pthreads = pd.read_csv("blackscholes/medium/benchmark-gcc-pthreads-55.out")[:54]
# blackscholes_ff = pd.read_csv("blackscholes/medium/benchmark-gcc-ff-55.out")[:54]
# blackscholes_openmp = pd.read_csv("blackscholes/medium/benchmark-gcc-openmp-55.out")[:54]
# blackscholes_tbb = pd.read_csv("blackscholes/medium/benchmark-gcc-tbb-55.out")[:54]
# blackscholes_skepu_openmp = pd.read_csv("blackscholes/medium/benchmark-gcc-skepu-openmp-55.out")[:54]
# blackscholes_hybrid = pd.read_csv("blackscholes/medium/benchmark-gcc-skepu-hybrid.out")[:54]
# serial_time = 0.214505
# serial_energy = 18.271
# label = "medium"

# blackscholes_pthreads = pd.read_csv("blackscholes/large/benchmark-gcc-pthreads-55.out")[:54]
# blackscholes_ff = pd.read_csv("blackscholes/large/benchmark-gcc-ff-55.out")[:54]
# blackscholes_openmp = pd.read_csv("blackscholes/large/benchmark-gcc-openmp-55.out")[:54]
# blackscholes_tbb = pd.read_csv("blackscholes/large/benchmark-gcc-tbb-55.out")[:54]
# blackscholes_skepu_openmp = pd.read_csv("blackscholes/large/benchmark-gcc-skepu-openmp-55.out")[:54]
# blackscholes_hybrid = pd.read_csv("blackscholes/large/benchmark-gcc-skepu-hybrid.out")[:54]
# serial_time = 0.673967
# serial_energy = 57.3285
# label = "large"

blackscholes_pthreads = pd.read_csv("blackscholes/native/benchmark-gcc-pthreads-55.out")[:54]
blackscholes_ff = pd.read_csv("blackscholes/native/benchmark-gcc-ff-55.out")[:54]
blackscholes_openmp = pd.read_csv("blackscholes/native/benchmark-gcc-openmp-55.out")[:54]
blackscholes_tbb = pd.read_csv("blackscholes/native/benchmark-gcc-tbb-55.out")[:54]
blackscholes_skepu_openmp = pd.read_csv("blackscholes/native/benchmark-gcc-skepu-openmp-55.out")[:54]
blackscholes_hybrid = pd.read_csv("blackscholes/native/benchmark-gcc-skepu-hybrid.out")[:54]
serial_time = 76.1651
serial_energy = 7652.32
label = "native"
```

```python
plt.figure(figsize=(10, 5))
plt.plot(blackscholes_ff.Threads,blackscholes_ff.JoulesCpuAll/ blackscholes_ff.Time,label = "ff")
plt.plot(blackscholes_pthreads.Threads,blackscholes_pthreads.JoulesCpuAll/ blackscholes_pthreads.Time,label = "pthreads")
plt.plot(blackscholes_tbb.Threads,blackscholes_tbb.JoulesCpuAll/ blackscholes_tbb.Time,label = "tbb")
plt.plot(blackscholes_skepu_openmp.Threads,blackscholes_skepu_openmp.JoulesCpuAll/ blackscholes_skepu_openmp.Time,label = "skepu-openmp")
plt.plot(blackscholes_openmp.Threads,blackscholes_openmp.JoulesCpuAll/ blackscholes_openmp.Time,label = "openmp")
plt.axhline(y=135, color='black', linestyle='-', label="PL_1")
plt.xlim([1,55])
plt.axhline(y=250, color='r', linestyle='-', label="PL_2")
plt.xlabel("Number of threads")
plt.ylabel("Power (Watts)")
plt.grid()
plt.legend()
plt.savefig('blackschools_'+label+'_power')
plt.show()
```

```python
fig, ax = plt.subplots(2,figsize=(10, 12))

ax[0].plot(blackscholes_pthreads.Threads, serial_time / blackscholes_pthreads.Time,label = "pthreads")
ax[0].plot(blackscholes_ff.Threads, serial_time / blackscholes_ff.Time, label = "ff")
ax[0].plot(blackscholes_openmp.Threads, serial_time / blackscholes_openmp.Time, label = "openmp")
ax[0].plot(blackscholes_tbb.Threads, serial_time / blackscholes_tbb.Time, label = "tbb")
ax[0].plot(blackscholes_skepu_openmp.Threads, serial_time / blackscholes_skepu_openmp.Time, label = "skepu-openmp")
ax[0].set_xlabel("Number of threads")
ax[0].set_ylabel("Speedup")
ax[0].set_xlim([1,55])
ax[0].grid()
ax[0].legend()

ax[1].plot(blackscholes_pthreads.Threads, (serial_energy - blackscholes_pthreads.PCM_energy) / serial_energy *100,label = "pthreads")
ax[1].plot(blackscholes_ff.Threads, (serial_energy - blackscholes_ff.PCM_energy) / serial_energy *100, label = "ff")
ax[1].plot(blackscholes_openmp.Threads, (serial_energy - blackscholes_openmp.PCM_energy) / serial_energy *100, label = "openmp")
ax[1].plot(blackscholes_tbb.Threads, (serial_energy - blackscholes_tbb.PCM_energy) / serial_energy *100, label = "tbb")
ax[1].plot(blackscholes_skepu_openmp.Threads, (serial_energy - blackscholes_skepu_openmp.PCM_energy) / serial_energy *100, label = "skepu-openmp")
ax[1].set_xlabel("Number of threads")
ax[1].set_ylabel("Energy saving (%)")
ax[1].set_xlim([1,55])
ax[1].grid()
ax[1].legend()

fig.savefig('blackschools_'+label+'_overall')
```

```python
fig, ax = plt.subplots(2,2,figsize=(18, 10))

openmp_vs_skepu_openmp = (blackscholes_skepu_openmp - blackscholes_openmp) / blackscholes_openmp * 100
openmp_vs_skepu_openmp[["Time", "PCM_energy"]].to_csv("temp_data", sep='\t')
ax[0,0].scatter(openmp_vs_skepu_openmp.Time, openmp_vs_skepu_openmp.PCM_energy)
ax[0,0].scatter(openmp_vs_skepu_openmp.Time.mean(), openmp_vs_skepu_openmp.PCM_energy.mean())
ax[0,0].annotate("mean",(openmp_vs_skepu_openmp.Time.mean(), openmp_vs_skepu_openmp.PCM_energy.mean()))
ax[0,0].axhline(0, color='black')
ax[0,0].axvline(0, color='black')
ax[0,0].set_xlabel('Time (%)')
ax[0,0].set_ylabel('Energy (%)')
ax[0,0].grid()
ax[0,0].set_title('SkePU_openmp')

for i in range(54):
    ax[0,0].annotate(i+1, (openmp_vs_skepu_openmp.Time[i], openmp_vs_skepu_openmp.PCM_energy[i]))
    
openmp_vs_Pthreads = (blackscholes_pthreads - blackscholes_openmp) / blackscholes_openmp * 100
openmp_vs_Pthreads[["Time", "PCM_energy"]].to_csv("temp_data", sep='\t')
ax[0,1].scatter(openmp_vs_Pthreads.Time, openmp_vs_Pthreads.PCM_energy)
ax[0,1].scatter(openmp_vs_Pthreads.Time.mean(), openmp_vs_Pthreads.PCM_energy.mean())
ax[0,1].annotate("mean",(openmp_vs_Pthreads.Time.mean(), openmp_vs_Pthreads.PCM_energy.mean()))
ax[0,1].axhline(0, color='black')
ax[0,1].axvline(0, color='black')
ax[0,1].set_xlabel('Time (%)')
ax[0,1].set_ylabel('Energy (%)')
ax[0,1].grid()
ax[0,1].set_title('Pthreads')

for i in range(54):
    ax[0,1].annotate(i+1, (openmp_vs_Pthreads.Time[i], openmp_vs_Pthreads.PCM_energy[i]))
    

openmp_vs_tbb = (blackscholes_tbb - blackscholes_openmp) / blackscholes_openmp * 100
openmp_vs_tbb[["Time", "PCM_energy"]].to_csv("temp_data", sep='\t')
ax[1,0].scatter(openmp_vs_tbb.Time, openmp_vs_tbb.PCM_energy)
ax[1,0].scatter(openmp_vs_tbb.Time.mean(), openmp_vs_tbb.PCM_energy.mean())
ax[1,0].annotate("mean",(openmp_vs_tbb.Time.mean(), openmp_vs_tbb.PCM_energy.mean()))
ax[1,0].axhline(0, color='black')
ax[1,0].axvline(0, color='black')
ax[1,0].set_xlabel('Time (%)')
ax[1,0].set_ylabel('Energy (%)')
ax[1,0].grid()
ax[1,0].set_title('TBB')

for i in range(54):
    ax[1,0].annotate(i+1, (openmp_vs_tbb.Time[i], openmp_vs_tbb.PCM_energy[i]))
    

openmp_vs_ff = (blackscholes_ff - blackscholes_openmp) / blackscholes_openmp * 100
openmp_vs_ff[["Time", "PCM_energy"]].to_csv("temp_data", sep='\t')
ax[1,1].scatter(openmp_vs_ff.Time, openmp_vs_ff.PCM_energy)
ax[1,1].scatter(openmp_vs_ff.Time.mean(), openmp_vs_ff.PCM_energy.mean())
ax[1,1].annotate("mean",(openmp_vs_ff.Time.mean(), openmp_vs_ff.PCM_energy.mean()))
ax[1,1].axhline(0, color='black')
ax[1,1].axvline(0, color='black')
ax[1,1].set_xlabel('Time (%)')
ax[1,1].set_ylabel('Energy (%)')
ax[1,1].grid()
ax[1,1].set_title('FastFlow')

for i in range(54):
    ax[1,1].annotate(i+1, (openmp_vs_ff.Time[i], openmp_vs_ff.PCM_energy[i]))
    
    
fig.savefig('blackschools_'+label+'_proportion')

#openmp_vs_skepu_openmp[["Time", "PCM_energy"]].mean()
```
